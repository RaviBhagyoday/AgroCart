```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_odisha=df[df['State_Name']=='Odisha']
df_odisha.head()
```

```python
df_odisha['District_Name'].unique()
```

```python
df_odisha.isnull().any()
```

```python
df_odisha.count()
```

```python
df_odisha[df_odisha['Production'].isnull()]
```

```python
df_odisha=df_odisha.dropna()
```

```python
df_odisha['Production'].mean()
```

```python
df_nocropyear=df_odisha.drop(['Crop_Year'],axis=1)
max_prod=df_nocropyear.groupby(['District_Name','Season']).max()
```

```python
max_prod1= max_prod[:30]

fig = plt.figure(figsize=(25,25))
ax = max_prod1.plot.bar()
plt.xticks()
plt.show()
```

```python
max_prod['Production'].max()
```

```python
max_prod[max_prod['Production']==715000.0]
```

```python
min_prod=df_nocropyear.groupby(['District_Name','Season']).min()
min_prod['Production'].min()
```

```python
min_prod[min_prod['Production']==0.1]
```

```python
min_prod1= min_prod[:20]

fig2 = plt.figure(figsize=(25,25))
ax = min_prod1.plot.bar()
plt.xticks()
plt.show()
```

```python
crop_type_count = df_odisha.groupby('Crop').count()['District_Name']
fig1 = plt.figure(figsize=(20,20))
ax = crop_type_count.sort_values(ascending=False).plot.bar()
for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x()+0.15, p.get_height()+1),fontsize=11)
plt.xlabel('Crop Type',fontsize = 20)
plt.xticks(rotation = 60)
plt.ylabel('Production in Districts',fontsize = 20)
plt.show()
```
